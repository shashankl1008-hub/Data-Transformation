2. Ensures Consistent Data Format

5. Makes Data Ready for Analysis

Ans Q 4--

1. Mean/Median Imputation

Replace missing Age values with the mean or median age of the dataset.

**When to Use**


When only a small percentage of values are missing.
When Age is an important feature and you do not want to lose records.
Median is preferred if there are outliers (e.g., ages 5, 20, 25, 30, 90).
Advantages
Simple and quick.
Preserves all records.
Disadvantages
Can reduce data variability.
May introduce bias if many values are missing.

Ans Q 8--


Z-Score Standardization is preferred over Min-Max Scaling when outliers exist because Min-Max Scaling uses the minimum and maximum values of the dataset, making it highly sensitive to extreme values. Outliers can compress most data points into a very small range. Z-Score Standardization uses the mean and standard deviation, which helps preserve the relative distribution of the data and reduces the impact of outliers.
